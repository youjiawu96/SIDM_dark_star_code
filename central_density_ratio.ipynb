```python
import Structure_Solver_property_return as slv
import pandas as pd
import numpy as np
```

```python
log_ratio = np.arange(-7.0,-2.0,0.2)
solar_mass = np.array([10,50,100,500])
AU = 1.496*10**11
R_guess = 10**11
for mass in solar_mass:
    L_DM = np.zeros(len(log_ratio))
    L_final = np.zeros(len(log_ratio))
    T_eff = np.zeros(len(log_ratio))
    R_photo = np.zeros(len(log_ratio))
    rho_cd = np.zeros(len(log_ratio))
    rho_cg = np.zeros(len(log_ratio))
    dmfrac = np.zeros(len(log_ratio))
    for i in range(len(log_ratio)):
        L_DM[i], L_final[i], T_eff[i], R_photo[i], rho_cd[i], rho_cg[i], dmfrac[i] = slv.properties(mass, 10**(log_ratio[i]),R_guess)
        R_guess = R_photo[i]*AU
    data = np.vstack((L_DM,L_final,T_eff,R_photo,rho_cd,rho_cg, dmfrac))
    data = np.transpose(data)
    data_df = pd.DataFrame(data = data, columns = ["L_DM","L_final","T_eff", "R_photo", "rho_cd", "rho_cg", "dmfrac"])
    data_df.to_hdf('central_density.h5',key='solar'+str(mass),mode='a')
    
```
