```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
upperlimit = np.loadtxt('upper.csv',delimiter=',')
lowerlimit = np.loadtxt('lower.csv',delimiter=',')
exclusion = np.loadtxt('exclusion.csv',delimiter=',')
upperlimit = 10**upperlimit
lowerlimit = 10**lowerlimit
exclusion = 10**exclusion
```

```python
data = np.loadtxt('scalar_exclude_lim.txt', delimiter=' ')
DMMass = data[:,0]
fav1 = data[:,1]
fav2 = data[:,2]
```

```python
cluster_1 = np.loadtxt("one_SIDM_cluster.txt", delimiter = '\t')
cluster_p1 = np.loadtxt("pointone_SIDM_cluster.txt", delimiter = '\t')
```

```python
height = np.loadtxt("CMB_constraint.txt", delimiter = '\t')
m_psi = 10**np.arange(2,3,0.01)
m_phi = 10**np.arange(-3,-1,0.01)
X, Y = np.meshgrid(m_psi, m_phi)
```

```python
plt.style.use('az-paper-twocol.mplstyle')
fig = plt.figure(figsize=(10,8))
plt.fill_between(upperlimit[:,1],upperlimit[:,0],color='silver')
plt.fill_between(lowerlimit[:,1],lowerlimit[:,0],color='white')

# plt.plot(DMMass,mediator1,ls = '--', color = 'purple')
# plt.plot(DMMass,mediator2,'b--')

plt.plot(DMMass,fav1,color = 'purple')
plt.plot(DMMass,fav2,color = 'b')
# plt.plot(DMMass,fav3,color = 'r')

# plt.plot(exclusion[:,1],exclusion[:,0],color='magenta')
# plt.fill_between(exclusion[:,1],exclusion[:,0],color='magenta',alpha=0.3)

# plt.plot(cluster_1[:,0], cluster_1[:,1], color = 'red')
plt.plot(cluster_p1[:,0], cluster_p1[:,1], color = 'orange')
plt.fill_between(cluster_p1[:,0], cluster_p1[:,1], color = 'orange', alpha=0.3)

plt.contourf(X, Y, height, levels = [1.0, 10000.0],hatches = '/', colors = 'none')
plt.contour(X, Y, height, levels = [1.0],colors = 'k', linestyle = 'solid')

# plt.plot(10**CMBdata[:,0], 10**CMBdata[:,1],ls='-.', color = 'magenta', linewidth = 5.0)

# plt.plot(DMMass,fav4,color = 'purple',ls='-.')
# plt.plot(DMMass,fav5,color = 'b',ls='-.')
# plt.plot(DMMass,fav6,color = 'r',ls='-.')

# plt.hlines([fav1,fav2,fav3],colors = ['purple','b','r'], xmin = min(lowerlimit[:,0]), xmax = 10**4)

plt.text(0.01,1.1*10**(-4),s = r'$(\sigma_{\rm SI}/m_{\rm \psi})_{\rm dwarf}=0.1\,{\rm cm^2/g}-10\,{\rm cm^2/g}$',fontsize = 26, rotation = 37)
plt.text(2,0.0015,s='favored by constraining\n'+r'$\big\uparrow\lambda_{\rm decay}<1{\rm AU}$'+' at '+r'$g_{\rm Y} = 8 \times 10^{-5}\big\uparrow$',color='purple', fontsize=26, rotation=26, rotation_mode='anchor')
plt.text(2,0.015,s='favored by constraining\n'+r'$\big\uparrow\lambda_{\rm decay}<1{\rm AU}$'+' at '+r'$g_{\rm Y}= 10^{-5}\big\uparrow$',color='blue', fontsize=26, rotation=24, rotation_mode='anchor')
plt.text(0.1, 3*10**(-4), s=r'$(\sigma_{\rm SI}/m_{\rm \psi})_{\rm cluster} > 0.1\,{\rm cm^2/g}$', fontsize = 26)
# plt.text(8,0.1,s='favored by constraining\n'+r'$\big\uparrow\lambda_{\rm decay}<5{\rm AU}$'+' at '+r'$\epsilon = 10^{-11}\big\uparrow$',color='red', fontsize=22, rotation=22, rotation_mode='anchor')

# plt.text(100,0.02,s=r'$\downarrow$'+'Xenon1T exclusion limit at '+r'$\epsilon = 10^{-10}\downarrow$',color='blue',fontsize=22)
# plt.text(100,0.13,s=r'$\downarrow$'+'Xenon1T exclusion limit at '+r'$\epsilon = 10^{-9}\downarrow$',color='purple',fontsize=22)
plt.xscale('log')
plt.yscale('log')
plt.xlim(min(lowerlimit[:,1]),10**3)
plt.ylim(10**(-4), 1)
# plt.ylim(10**(-3),1)
plt.xlabel(r'DM Particle Mass ($m_\psi$) [GeV]')
plt.ylabel(r'Scalar Mediator Mass ($m_\phi$) [GeV]')
plt.tight_layout()
fig.savefig('constrain_mod.pdf')
# fig.savefig('constrain_mod.png')
```
