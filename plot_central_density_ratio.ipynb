```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('az-paper-onecol.mplstyle')
```

```python
wimp_ds = np.loadtxt("Fig4.txt", delimiter = ',') # columns: log10(mdm/m), Ldm/Ltot, log10(Ltot/Lsun), Teff, radius in AU, log10(rho_dm), log10(rho_gas)
print(wimp_ds)
solar10 = pd.read_hdf('central_density.h5','solar10')
solar50 = pd.read_hdf('central_density.h5','solar50')
solar100 = pd.read_hdf('central_density.h5','solar100')
solar500 = pd.read_hdf('central_density.h5','solar500')
```

```python
log_ratio = np.arange(-7.0,-2.0,0.2)
ratio = 10**log_ratio
fig, az = plt.subplots(3,2,figsize = (12,12),sharex=True)
az[0,0].plot(solar10['dmfrac'], solar10['L_DM']/solar10['L_final'],'r-',label = r'$M_* = 10 M_{\odot}$')
az[0,0].plot(solar50['dmfrac'], solar50['L_DM']/solar50['L_final'],'b--',label = r'$M_* = 50 M_{\odot}$')
az[0,0].plot(solar100['dmfrac'], solar100['L_DM']/solar100['L_final'],'m:',label = r'$M_* = 100 M_{\odot}$')
az[0,0].plot(solar500['dmfrac'], solar500['L_DM']/solar500['L_final'],'k-.',label = r'$M_* = 500 M_{\odot}$')
az[0,0].scatter(10**wimp_ds[0,0],wimp_ds[0,1],c = 'r', marker = 'o', s =400, alpha = 0.5)
az[0,0].scatter(10**wimp_ds[1,0],wimp_ds[1,1],c= 'b', marker = '^', s = 400, alpha = 0.5)
az[0,0].scatter(10**wimp_ds[2,0],wimp_ds[2,1], c = 'm', marker = 's', s = 400, alpha = 0.5)
az[0,0].scatter(10**wimp_ds[3,0],wimp_ds[3,1], c = 'k', marker = '*', s = 400, alpha = 0.5)
az[0,0].axhline(y=0.5,linestyle='--',color='k')
az[0,0].set_xscale('log')
az[0,0].set_ylabel(r'$L_{\rm DM}/L_{\rm tot}$')
az[0,0].set_xlim(10**(-7.0),10**(-2.0))

az[1,0].plot(solar10['dmfrac'], solar10['L_final'],'r-',label = r'$M_* = 10 M_{\odot}$')
az[1,0].plot(solar50['dmfrac'], solar50['L_final'],'b--',label = r'$M_* = 50 M_{\odot}$')
az[1,0].plot(solar100['dmfrac'], solar100['L_final'],'m:',label = r'$M_* = 100 M_{\odot}$')
az[1,0].plot(solar500['dmfrac'], solar500['L_final'],'k-.',label = r'$M_* = 500 M_{\odot}$')
az[1,0].scatter(10**wimp_ds[0,0],10**wimp_ds[0,2],c = 'r', marker = 'o', s = 400, alpha = 0.5)
az[1,0].scatter(10**wimp_ds[1,0],10**wimp_ds[1,2],c= 'b', marker = '^', s = 400, alpha = 0.5)
az[1,0].scatter(10**wimp_ds[2,0],10**wimp_ds[2,2], c = 'm', marker = 's', s = 400, alpha = 0.5)
az[1,0].scatter(10**wimp_ds[3,0],10**wimp_ds[3,2], c = 'k', marker = '*', s = 400, alpha = 0.5)
az[1,0].set_xscale('log')
az[1,0].set_yscale('log')
az[1,0].set_ylabel(r'$L_{\rm tot}\,[L_\odot]$')
az[1,0].set_xlim(10**(-7.0),10**(-2.0))

az[2,0].plot(solar10['dmfrac'], solar10['T_eff'],'r-',label = r'$M_* = 10 M_{\odot}$')
az[2,0].plot(solar50['dmfrac'], solar50['T_eff'],'b--',label = r'$M_* = 50 M_{\odot}$')
az[2,0].plot(solar100['dmfrac'], solar100['T_eff'],'m:',label = r'$M_* = 100 M_{\odot}$')
az[2,0].plot(solar500['dmfrac'], solar500['T_eff'],'k-.',label = r'$M_* = 500 M_{\odot}$')
az[2,0].scatter(10**wimp_ds[0,0],wimp_ds[0,3],c = 'r', marker = 'o', s = 400, alpha = 0.5)
az[2,0].scatter(10**wimp_ds[1,0],wimp_ds[1,3],c= 'b', marker = '^', s = 400, alpha = 0.5)
az[2,0].scatter(10**wimp_ds[2,0],wimp_ds[2,3], c = 'm', marker = 's', s = 400, alpha = 0.5)
az[2,0].scatter(10**wimp_ds[3,0],wimp_ds[3,3], c = 'k', marker = '*', s = 400, alpha = 0.5)
az[2,0].set_xscale('log')
az[2,0].set_ylabel(r'$T_{\rm eff}\,[\rm K]$')
az[2,0].set_xlim(10**(-7.0),10**(-2.0))
az[2,0].axes.xaxis.set_ticks([10**(-7),10**(-6),10**(-5),10**(-4),10**(-3),10**(-2)])
az[2,0].set_xlabel('Dark Matter Mass Fraction')

az[0,1].plot(solar10['dmfrac'], solar10['R_photo'],'r-',label = r'$M = 10 M_{\odot}$')
az[0,1].plot(solar50['dmfrac'], solar50['R_photo'],'b--',label = r'$M = 50 M_{\odot}$')
az[0,1].plot(solar100['dmfrac'], solar100['R_photo'],'m:',label = r'$M = 100 M_{\odot}$')
az[0,1].plot(solar500['dmfrac'], solar500['R_photo'],'k-.',label = r'$M = 500 M_{\odot}$')
az[0,1].scatter(10**wimp_ds[0,0],wimp_ds[0,4],c = 'r', marker = 'o', s = 400, alpha = 0.5)
az[0,1].scatter(10**wimp_ds[1,0],wimp_ds[1,4],c= 'b', marker = '^', s = 400, alpha = 0.5)
az[0,1].scatter(10**wimp_ds[2,0],wimp_ds[2,4], c = 'm', marker = 's', s = 400, alpha = 0.5)
az[0,1].scatter(10**wimp_ds[3,0],wimp_ds[3,4], c = 'k', marker = '*', s = 400, alpha = 0.5)
az[0,1].set_xscale('log')
az[0,1].set_ylabel(r'$R_*\,[{\rm AU}]$')
az[0,1].set_xlim(10**(-7.0),10**(-2.0))
az[0,1].legend()

az[1,1].plot(solar10['dmfrac'], solar10['rho_cd'],'r-',label = r'$M_* = 10 M_{\odot}$')
az[1,1].plot(solar50['dmfrac'], solar50['rho_cd'],'b--',label = r'$M_* = 50 M_{\odot}$')
az[1,1].plot(solar100['dmfrac'], solar100['rho_cd'],'m:',label = r'$M_* = 100 M_{\odot}$')
az[1,1].plot(solar500['dmfrac'], solar500['rho_cd'],'k-.',label = r'$M_* = 500 M_{\odot}$')
az[1,1].scatter(10**wimp_ds[0,0],10**wimp_ds[0,5],c = 'r', marker = 'o', s = 400, alpha = 0.5)
az[1,1].scatter(10**wimp_ds[1,0],10**wimp_ds[1,5],c= 'b', marker = '^', s = 400, alpha = 0.5)
az[1,1].scatter(10**wimp_ds[2,0],10**wimp_ds[2,5], c = 'm', marker = 's', s = 400, alpha = 0.5)
az[1,1].scatter(10**wimp_ds[3,0],10**wimp_ds[3,5], c = 'k', marker = '*', s = 400, alpha = 0.5)
az[1,1].set_xscale('log')
az[1,1].set_yscale('log')
az[1,1].set_ylabel(r'$\rho_{c,{\rm DM}}\,[{\rm GeV/cm^3}]$')
az[1,1].set_xlim(10**(-7.0),10**(-2.0))
az[1,1].set_yticks([10**12, 10**13, 10**14])

az[2,1].plot(solar10['dmfrac'], solar10['rho_cg'],'r-',label = r'$M_* = 10 M_{\odot}$')
az[2,1].plot(solar50['dmfrac'], solar50['rho_cg'],'b--',label = r'$M_* = 50 M_{\odot}$')
az[2,1].plot(solar100['dmfrac'], solar100['rho_cg'],'m:',label = r'$M_* = 100 M_{\odot}$')
az[2,1].plot(solar500['dmfrac'], solar500['rho_cg'],'k-.',label = r'$M_* = 500 M_{\odot}$')
az[2,1].scatter(10**wimp_ds[0,0],10**wimp_ds[0,6],c = 'r', marker = 'o', s = 400, alpha = 0.5)
az[2,1].scatter(10**wimp_ds[1,0],10**wimp_ds[1,6],c= 'b', marker = '^', s = 400, alpha = 0.5)
az[2,1].scatter(10**wimp_ds[2,0],10**wimp_ds[2,6], c = 'm', marker = 's', s = 400, alpha = 0.5)
az[2,1].scatter(10**wimp_ds[3,0],10**wimp_ds[3,6], c = 'k', marker = '*', s = 400, alpha = 0.5)
az[2,1].set_xscale('log')
az[2,1].set_yscale('log')
az[2,1].set_ylabel(r'$\rho_{c, {\rm gas}}\,[{\rm GeV/cm^3}]$')
az[2,1].set_xlim(10**(-7.0),10**(-2.0))
az[2,1].axes.xaxis.set_ticks([10**(-7),10**(-6),10**(-5),10**(-4),10**(-3),10**(-2)])
az[2,1].set_xlabel('Dark Matter Mass Fraction')

plt.subplots_adjust(wspace = 0.25,hspace = 0.0)
fig.savefig("fd_combined_original.pdf")
```
