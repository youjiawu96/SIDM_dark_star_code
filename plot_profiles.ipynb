```python
import Structure_Solver_prof_return as slv
import numpy as np
```

```python
rho_d, rho_g, T, P, r, photo = slv.profiles(10,1.5*10**(-3))
print(np.shape(rho_g))
print(np.shape(rho_d))
print(np.shape(r))
```

```python
import matplotlib.pyplot as plt
plt.style.use('az-paper-twocol.mplstyle')
AU = 1.496*10**13 #Au in cm
g = 5.61*10**23 # g in GeV
```

```python
cdm_prof = np.loadtxt("DSprofile.csv", delimiter=',')
cdm_r = cdm_prof[:,2]/AU
cdm_rhog = cdm_prof[:,3]*g
cdm_rhoD = cdm_prof[:,1]*g
cdm_T = cdm_prof[:,5]
cdm_P = cdm_prof[:,4]
```

```python
length = len(rho_g)
r = r[:length]
fig, ax = plt.subplots(3, sharex=True, figsize=(10,12))
ax[0].plot(r,rho_d,'r-',label = "DM density profile")
ax[0].plot(r,rho_g,'b-',label = "gas density profile")
ax[0].plot(cdm_r, cdm_rhoD, 'r--')
ax[0].plot(cdm_r, cdm_rhog, 'b--')
ax[0].set_ylabel('Density '+r'${\rm [GeV/cm^3]}$')
ax[0].set_yscale('log')
ax[0].axvline(x=3.16, ls = '-.')
ax[0].legend()

ax[1].plot(r,T,'k-', label="SIDM DS")
ax[1].plot(cdm_r, cdm_T,'k--', label="CDM DS")
ax[1].axvline(x=3.16,ls='-.')
ax[1].set_ylabel('Temperature [K]')
ax[1].set_yscale('log')
ax[1].legend()

ax[2].plot(r,P,'k-')
ax[2].plot(cdm_r, cdm_P,'k--')
ax[2].axvline(x=3.16,ls='-.')
ax[2].set_ylabel('Gas Pressure [Pa]')
ax[2].set_yscale('log')
ax[2].set_xlabel(r'$r$ [AU]')

plt.subplots_adjust(hspace = 0.0)
plt.xlim(0.0,3.4)
fig.savefig('profiles.pdf')
```

```python
plt.figure()
photo = photo[:length]
plt.plot(r, photo)
plt.yscale('log')
plt.show()
print()
```
